```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean_absolute_error(y, yp):
    mae = sum(abs(y - yp))/len(y)
    return mae

def mean_square_error(y, yp):
    mse = sum( (y-yp) ** 2 ) /len(y)
    return mse
```
